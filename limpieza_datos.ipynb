## Barrio

La descripcion del barrio en las publicaciones no es consistente, puede aparecer luego de la coma: Capital Federal
o Centro o un barrio como Palermo. Además tiene barrios como subbarrios en el primer termino. 

Se define funcion para manejar estas inconsistencias 

```python
df['barrio']
```

```python
def limpiar_barrio(texto):
    # Busca la última ocurrencia de " en "
    idx = texto.lower().rfind(" en ")
    if idx != -1:
        resto = texto[idx + 4:]  # Salta " en "
        partes = [p.strip() for p in resto.split(",")]
        if len(partes) == 2:
            # Caso 1: Si el segundo miembro es "Capital Federal", elegir el primero
            if partes[1].lower() == "capital federal":
                return partes[0]
            # Caso 2: Si el segundo miembro es "Centro", elegir el primero
            elif partes[1].lower() == "centro":
                return partes[0]
            # Caso 3: Si el segundo miembro es "Palermo", elegir el segundo
            elif partes[1].lower() == "palermo":
                return partes[1]
            # Caso 4: Si el segundo miembro es "Barrio Norte", elegir el segundo
            elif partes[1].lower() == "barrio norte":
                return partes[1]
            # Caso 5: Si el segundo miembro es "Floresta" y el primero es "Monte Castro", elegir el primero
            elif partes[1].lower() == "floresta" and partes[0].lower() == "monte castro":
                return partes[0]
            # Caso 6: Si el segundo miembro es "Floresta" y el primero es "Parque Avellaneda", elegir el primero
            elif partes[1].lower() == "floresta" and partes[0].lower() == "parque avellaneda":
                return partes[0]
            # Caso 7: Si el segundo miembro es "Villa Urquiza" y el primero es "Parque Chas", elegir el primero
            elif partes[1].lower() == "villa urquiza" and partes[0].lower() == "parque chas":
                return partes[0]
            # Caso por defecto: elegir el segundo miembro
            else:
                return partes[1]
        else:
            return partes[0]  # en caso que haya solo un miembro elige el primero
    return texto
```

```python
df['barrio_caba'] = df['barrio'].apply(limpiar_barrio)
df[['barrio', 'barrio_caba']]
```

## Características

```python
import re

def extraer_caracteristicas(texto):
    m2 = None
    antiguedad = None
    dorm = None

    if not isinstance(texto, str):
        return {"m2_cub": m2, "antig": antiguedad, "dorm": dorm}

    # m2 cubiertos
    m2_match = re.search(r'(\d+[\.,]?\d*)\s*m[²2]?\s*cub', texto.replace(" ", ""), re.IGNORECASE)
    if m2_match:
        m2 = m2_match.group(1).replace(',', '.')

    # antigüedad
    if re.search(r"a estrenar", texto, re.IGNORECASE):
        antiguedad = 0
    else:
        antiguedad_match = re.search(r'(\d+)\s*años?', texto, re.IGNORECASE)
        if antiguedad_match:
            antiguedad = int(antiguedad_match.group(1))

    # dormitorios
    dorm_match = re.search(r'(\d+)\s*dorm(itorios)?', texto, re.IGNORECASE)
    if dorm_match:
        dorm = int(dorm_match.group(1))
    elif re.search(r'monoam', texto, re.IGNORECASE):
        dorm = 1

    return {"m2_cub": m2, "antig": antiguedad, "dorm": dorm}
```

Extraer características del departamento

```python
df['caracteristicas'] = df['caracteristicas'].astype(str)  # por si hay valores no string
# Aplicar la función extraer_caracteristicas a cada fila
caracteristicas = df['caracteristicas'].apply(extraer_caracteristicas)
# Convertir el resultado en un DataFrame
caracteristicas_df = pd.DataFrame(caracteristicas.tolist())
# Agregar las nuevas columnas al DataFrame original
df['m2_cub'] = caracteristicas_df['m2_cub']
df['antig'] = caracteristicas_df['antig']
df['dorm'] = caracteristicas_df['dorm']
df.head()
```

Reemplazo nulos en m2_cub con valores m2 si aparecen en descripcion_1

```python
# Extraer metros cuadrados de la descripción2 cuando m2_cub es None
def extract_m2(descripcion):
    if pd.isna(descripcion):
        return None

    # Busca patrones como "50 m2", "50m2", "50M2", "50 metros cuadrados", "50 m²"
    pattern = r'(\d+(?:[\.,]\d+)?)\s*(m2|m²|metros cuadrados)'
    match = re.search(pattern, descripcion.lower())
    if match:
        return float(match.group(1).replace(',', '.'))
    return None

# Aplicar la función a las filas donde m2_cub es None
mask = df['m2_cub'].isna()
df.loc[mask, 'm2_cub'] = df.loc[mask, 'descripcion_1'].apply(extract_m2)
```

```python
# Aplicar la función a las filas donde m2_cub es None
mask = df['m2_cub'].isna()
df.loc[mask, 'm2_cub'] = df.loc[mask, 'descripcion_2'].apply(extract_m2)
```

## Antiguedad

```python
print(len(df[df['antig'].isna()]))
df[df['antig'].isna()].head()
```

```python
# Extraer antigüedad de las descripciones cuando antig es None
def extract_antig(descripcion):
    if pd.isna(descripcion):
        return None
    desc = descripcion.lower()
    # Primero verificar si es a estrenar
    if "a estrenar" in desc:
        return 0
    # Buscar número de años, permitiendo punto o coma después
    match = re.search(r'(\d+)\s*año[s]?[.,]?', desc)
    if match:
        return int(match.group(1))
    return None

# Aplicar la función a las filas donde antig es None, primero en descripcion_1 y luego en descripcion_2
mask = df['antig'].isna()
df.loc[mask, 'antig'] = df.loc[mask, 'descripcion_1'].apply(extract_antig)
mask = df['antig'].isna()
df.loc[mask, 'antig'] = df.loc[mask, 'descripcion_2'].apply(extract_antig)
```

```python
# Extraer dormitorios de las descripciones cuando dorm es None
def extract_dorm(descripcion):
    if pd.isna(descripcion):
        return None
    desc = descripcion.lower()

    # 1. Buscar si es monoambiente o mono
    if re.search(r'monoamb|mono ambiente|monoambiente|^mono\b', desc):
        return 1
    # 2. Buscar cantidad de dormitorios (ej: 2 dorm, 3 dormitorios)
    match = re.search(r'(\d+)\s*dorm(itorio|itorios)?', desc)
    if match:
        return int(match.group(1))
    # 3. Buscar cantidad de ambientes (ej: 2 ambientes, 3 ambiente, 2 amb)
    match = re.search(r'(\d+)\s*amb(iente(s)?)?', desc)
    if match:
        return int(match.group(1))
    # 4. Si aparece solo "ambiente" o "amb" (sin número), dorm = 1
    if re.search(r'\bamb(iente)?(s)?\b', desc):
        return 1
    
    return None

# Aplicar la función a las filas donde dorm es None, primero en descripcion_1 y luego en descripcion_2
mask = df['dorm'].isna()
df.loc[mask, 'dorm'] = df.loc[mask, 'descripcion_1'].apply(extract_dorm)
mask = df['dorm'].isna()
df.loc[mask, 'dorm'] = df.loc[mask, 'descripcion_2'].apply(extract_dorm)
```
